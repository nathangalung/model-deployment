```python
# please cut the dataset_all.parquet to be only use 1000 row per partition date unique value, just use pd.read_parquet('data/dataset_all/parquet')
import pandas as pd

# Load the dataset
df = pd.read_parquet('data/dataset_all.parquet')
# the column of the date is partition_date, just use based on that column unqiue values
unique_dates = df['partition_date'].unique()
# Limit to 1000 rows per unique date
limited_df = pd.concat([
    df[df['partition_date'] == date].head(3000) for date in unique_dates
])
# Save the limited dataset back to local
limited_df.to_parquet('data/dataset_all_sample.parquet', index=False)
```
